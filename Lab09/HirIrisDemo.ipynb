```python
# Hierarchical Clustering Lab Exercise: Iris Dataset

# Step 1: Import Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.datasets import load_iris

# Step 2: Load the Dataset
# Dataset Description:
# The Iris dataset consists of measurements of iris flowers:
# - Features: Sepal Length, Sepal Width, Petal Length, Petal Width
# - Target: Species (Setosa, Versicolor, Virginica)
# Dataset Source: Built-in sklearn dataset
# https://www.kaggle.com/datasets/uciml/iris

iris = load_iris()
X = pd.DataFrame(iris.data, columns=iris.feature_names)
X.head()

# Step 3: Select Features for Clustering
# Let's use all 4 features
X_selected = X

# Step 4: Standardize the Features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_selected)

# Step 5: Visualize Dendrogram
linked = linkage(X_scaled, method='ward')

plt.figure(figsize=(10,7))
dendrogram(linked)
plt.title('Hierarchical Clustering Dendrogram (Iris Dataset)')
plt.xlabel('Samples')
plt.ylabel('Distance')
plt.show()

# Step 6: Apply Agglomerative Clustering
# Based on dendrogram or known number of classes (3)
hc = AgglomerativeClustering(n_clusters=3, linkage='ward')
clusters = hc.fit_predict(X_scaled)

# Add cluster labels to a new dataframe
df = X.copy()
df['Cluster_HC'] = clusters

# Step 7: Visualize Clustering Results
# We will plot Petal Length vs Petal Width colored by cluster
plt.figure(figsize=(8,6))
sns.scatterplot(x=df['petal length (cm)'], y=df['petal width (cm)'], hue=df['Cluster_HC'], palette='tab10')
plt.title('Hierarchical Clustering on Iris Dataset')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.legend(title='Cluster Label')
plt.show()

# Step 8: Silhouette Score
if len(set(clusters)) > 1:
    silhouette_hc = silhouette_score(X_scaled, clusters)
    print(f"\nSilhouette Score (Hierarchical Clustering, 4 features): {silhouette_hc:.3f}")
else:
    print("\nSilhouette Score (Hierarchical Clustering, 4 features): Not applicable (only 1 cluster)")

# -----------------------------
# TODO: Student Task 4 (Optional: Try with only Petal features)
# -----------------------------

'''
# Step 9: Try Clustering using only Petal Length and Petal Width
X_petal = X[['petal length (cm)', 'petal width (cm)']]
X_petal_scaled = scaler.fit_transform(X_petal)

# Dendrogram for petal features
linked_petal = linkage(X_petal_scaled, method='ward')

plt.figure(figsize=(10,7))
dendrogram(linked_petal)
plt.title('Dendrogram (Petal Features Only)')
plt.xlabel('Samples')
plt.ylabel('Distance')
plt.show()

# Apply Agglomerative Clustering
hc_petal = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='ward')
clusters_petal = hc_petal.fit_predict(X_petal_scaled)

df['Cluster_Petal'] = clusters_petal

# 2D Visualization (Petal features)
plt.figure(figsize=(8,6))
sns.scatterplot(x=df['petal length (cm)'], y=df['petal width (cm)'], hue=df['Cluster_Petal'], palette='tab10')
plt.title('Hierarchical Clustering (Petal Features Only)')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.legend(title='Cluster Label')
plt.show()

# Silhouette Score for petal features
if len(set(clusters_petal)) > 1:
    silhouette_petal = silhouette_score(X_petal_scaled, clusters_petal)
    print(f\"\\nSilhouette Score (Hierarchical Clustering on Petal Features): {silhouette_petal:.3f}\")
else:
    print(\"\\nSilhouette Score (Hierarchical Clustering on Petal Features): Not applicable (only 1 cluster)\")
'''

# Step 10: Student Analysis Tasks
print("\n--- Student Analysis Tasks ---")
print("1. (Conceptual) What is the difference between 'ward', 'single', and 'complete' linkage methods? Try changing the linkage method to observe the change in the outputs.")
print("2. (Interpretation) How many clusters did you find? How good is the clustering according to the Silhouette Score?")
print("3. (Hands-on) Try changing the number of clusters (n_clusters=2,4) and observe the changes in silhouette score. What does that hyperparameter signify?")
print("4. (Hands-on) Uncomment the TODO section and try clustering using only Petal Length and Petal Width. How does the clustering quality change?")
```
